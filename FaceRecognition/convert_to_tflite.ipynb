# This notebook is used to convert the model to tflite format

TensorFlow Lite is designed to provide efficient and fast performance on embedded devices, with lower memory footprint and computational requirements compared to TensorFlow.

```python
import tensorflow as tf

#print(help(tf.lite.TFLiteConverter))


modelPath = 'models/best/model_best1.h5'
savePath = 'models/tflite/recognition.tflite'
model = tf.keras.models.load_model(modelPath)


converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
open(savePath, "wb").write(tflite_model)
```
