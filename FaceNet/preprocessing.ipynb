```python
import os
import json
import cv2
import matplotlib.pyplot as plt
#current directory
print(os.getcwd())
imagePath = "faces\\pupsjson\\"
imagePaths = []
faces = []
print(len(os.listdir(imagePath)))
for file in os.listdir(imagePath):
    if file.endswith(".json"):
        with open(imagePath+file) as f:
            data = json.load(f)
            imagePaths.append(data['asset']['path'].split(":")[-1])
            #print(data['asset']['path'].split(":")[-1])
            boundingBoxes = ""
            for i in range(len(data['regions'])):            
                if boundingBoxes == "":
                    boundingBoxes = [data['regions'][i]['boundingBox']]
                else:
                    boundingBoxes.append(data['regions'][i]['boundingBox'])
            #boundingBox = data['regions'][0]['boundingBox']
            faces.append(boundingBoxes)


fig, ax = plt.subplots(1,5, figsize=(20,20))
for i in range(5):
    image = cv2.imread(imagePaths[i])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    for j in range(len(faces[i])):
        coordinates = (faces[i][j]['left'], faces[i][j]['top'], faces[i][j]['width'], faces[i][j]['height'])
        x,y,w,h = coordinates
        topleft = (int(x),int(y))
        bottomright = (int(x)+int(w), int(y)+int(h))
        cv2.rectangle(image, (topleft), (bottomright), (0,255,0), 2) 
    ax[i].imshow(image)
    ax[i].axis("off")
plt.show()
fig, ax = plt.subplots(1,5, figsize=(20,20))
for i in range(5):
    i = i+5
    image = cv2.imread(imagePaths[i])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    for j in range(len(faces[i])):
        coordinates = (faces[i][j]['left'], faces[i][j]['top'], faces[i][j]['width'], faces[i][j]['height'])
        x,y,w,h = coordinates
        topleft = (int(x),int(y))
        bottomright = (int(x)+int(w), int(y)+int(h))
        cv2.rectangle(image, (topleft), (bottomright), (0,255,0), 2) 
    ax[i-5].imshow(image)
    ax[i-5].axis("off")
plt.show()

print(len(imagePaths))
```

```python
#make image fro boundingbox
def makeImage(image, boundingBox):
    x,y,w,h = boundingBox
    topleft = (int(x),int(y))
    bottomright = (int(x)+int(w), int(y)+int(h))
    image = image[topleft[1]:bottomright[1], topleft[0]:bottomright[0]]
    return image

savePath = "faces\\processed\\"
for i in range(len(imagePaths)):
    image = cv2.imread(imagePaths[i])
    for j in range(len(faces[i])):
        coordinates = (faces[i][j]['left'], faces[i][j]['top'], faces[i][j]['width'], faces[i][j]['height'])
        imageNew = makeImage(image, coordinates)
        cv2.imwrite(savePath+str(i)+"_"+str(j)+".jpg", imageNew)



```
